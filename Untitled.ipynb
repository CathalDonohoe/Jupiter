```python
import numpy as np  #import numpy
import pandas as panda  #import panda
from sklearn.model_selection import train_test_split
import tensorflow as tensor
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD,Adam
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten



# Neural networks.
import tensorflow.keras as kr

# Numerical arrays
import numpy as np

# Data frames.
import pandas as pd

# Plotting
import matplotlib.pyplot as plt

```

```python
myPPData = panda.read_csv("https://raw.githubusercontent.com/ianmcloughlin/2020A-machstat-project/master/dataset/powerproduction.csv")
myPPData.head(20)
```

```python
myPPData.describe()
```

```python
# Train a different model.
model = kr.models.Sequential()
model.add(kr.layers.Dense(50, input_shape=(1,), activation='sigmoid', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
model.add(kr.layers.Dense(1, activation='linear', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
model.compile(kr.optimizers.Adam(lr=0.001), loss='mean_squared_error')
```

```python
# Train the neural network on our training data.
model.fit(myPPData['speed'], myPPData['power'], epochs=500, batch_size = 500)
```

```python
# Plot the values.
plt.plot(myPPData['speed'],myPPData['power'], 'o');
plt.plot(myPPData['speed'],model.predict(myPPData['speed']), 'bx');
```
