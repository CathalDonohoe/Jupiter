```python
import numpy as np  #import numpy
import pandas as panda  #import panda
from sklearn.model_selection import train_test_split
import tensorflow as tensor
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD,Adam
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten



# Neural networks.
import tensorflow.keras as kr

# Numerical arrays
import numpy as np

# Data frames.
import pandas as pd

# Plotting
import matplotlib.pyplot as plt

```

```python
myPPData = panda.read_csv("https://raw.githubusercontent.com/ianmcloughlin/2020A-machstat-project/master/dataset/powerproduction.csv")
myPPData.head(20)
```

```python
myPPData.describe()
```

```python
# Build a simple model
inputs = kr.Input(shape=(1, ))

x1 = Dense(10, activation="relu")(inputs)
x2 = Dense(50, activation="relu")(x1)
x3 = Dense(50, activation="relu")(x2)
 
outputs = Dense(1)(x3)
model = kr.Model(inputs=inputs, outputs=outputs) 
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
# Train the neural network on our training data.

model.fit(myPPData.iloc[:,0].values, myPPData.iloc[:,1].values, epochs=30000, batch_size = 100)
```

```python
# Plot the values.
plt.plot(myPPData['speed'],myPPData['power'], 'o');
plt.plot(myPPData['speed'],model.predict(myPPData['speed']), 'bx');
```
